```python
from matplotlib import pyplot as plt
import numpy as np

fix, ax = plt.subplots(1,1)
x = np.linspace(.0001,1,101)
y = -(np.log(x))

ax.set(xlabel = "z", ylabel = "f(z)")

ax.plot(x, y)
```

```python
from matplotlib import pyplot as plt
import numpy as np

fix, ax = plt.subplots(1,1)
x = np.linspace(0,.99999,101)
y = -(np.log(1-x))

ax.set(xlabel = "z", ylabel = "g(z)")

ax.plot(x, y)
```
